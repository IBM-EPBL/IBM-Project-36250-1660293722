## 6. Find the outliers and replace them outliers

```python
def box_scatter(data, x, y):    
    fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(16,6))
    sns.boxplot(data=data, x=x, ax=ax1)
    sns.scatterplot(data=data, x=x,y=y,ax=ax2)
```

```python
box_scatter(data,'Rings','Age');
plt.tight_layout()
print(f"# of Bivariate Outliers: {len(data.loc[data['Rings'] < 4177])}")
```

```python
box_scatter(data,'Rings','Integer');
plt.tight_layout()
print(f"# of Bivariate Outliers: {len(data.loc[data['Age'] > 220000])}")
```

```python
box_scatter(data,'Rings','Age');
plt.tight_layout()
```

### Removing Outliers

```python
for i in data:
    if data[i].dtype=='int64' or data[i].dtypes=='float64':
        q1=data[i].quantile(0.25)
        q3=data[i].quantile(0.75)
        iqr=q3-q1
        upper=q3+1.5*iqr
        lower=q1-1.5*iqr
        data[i]=np.where(data[i] >upper, upper, data[i])
        data[i]=np.where(data[i] <lower, lower, data[i])
```

## 7. Check for Categorical columns and perform encoding.

```python
labelencoder = LabelEncoder()
data['Rings']= labelencoder.fit_transform(data['Rings'])
data['Age'] = labelencoder.fit_transform(data['Age'])
```
